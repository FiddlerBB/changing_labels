```python
# sample_path = 'sample'
# xls_path = 'sample/*.xls'
# new_path = 'sample_out'
# csv_path = f'{new_path}/*.csv'

output_folder = 'labels_front_cmnd_output'


def create_folder(new_path):
    if not os.path.exists(new_path):
        os.makedirs(new_path)
    else:
        shutil.rmtree(new_path)
        os.makedirs(new_path)

create_folder(output_folder)
```

```python
path = 'labels_cccd_cmnd/labels_front_cmnd/*.txt'
output_folder = 'labels_front_cmnd_output'
for file in glob.glob(path):
    name = os.path.basename(file)
    df = pd.read_csv(file, header=None, sep = ' ')
    df = df.replace({0: cmnd_dict})
    final_path = os.path.join(output_folder, name)
    df.to_csv(final_path, index=False, header=None, sep = ' ')
```
